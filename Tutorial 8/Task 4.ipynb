```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#load the image in
img = mpimg.imread('window.png')

# just extract one channel - this is red
imgBw = np.squeeze(img[:,:,0])

# get the size
npy, npx = imgBw.shape

plt.imshow(imgBw, cmap=plt.cm.gray)
plt.axis('off')
```

```python
diffx = np.diff(imgBw, axis=1)
diffy = np.diff(imgBw, axis=0)

diffx = diffx[0:npy-1,:]
diffy = diffy[:,0:npx-1]

diff = np.sqrt(np.power(diffx,2)+np.power(diffy,2))

plt.imshow(diff,cmap=plt.cm.gray)
plt.axis('off')
plt.colorbar()
```

```python
t = 0.1

edgeThresh = (diff > t).astype('int')

plt.imshow(edgeThresh,cmap=plt.cm.gray)
plt.axis('off')
```
