```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#load the image in
img = mpimg.imread('window.png')

# just extract one channel - this is red
imgBw = np.squeeze(img[:,:,0])

# get the size
npy, npx = imgBw.shape

plt.imshow(imgBw, cmap=plt.cm.gray)
plt.axis('off')
```

```python
fim = np.fft.fft2(imgBw)

fim = np.fft.fftshift(fim)

plt.figure()
plt.imshow(np.abs(fim))
# showing the colour scale
plt.colorbar()
# set the colour limits
plt.clim(0,1e3)
```

```python
#plotting the x lines
# make a separate copy of image to work on
fim2 = fim.copy()
# set everything in all except the first row to zero
fim2[1:npy,:] = 0
# set everything in the first row to zero except the first 61 points and the last 60
fim2[0,61:npx-60] = 0

x = np.fft.ifft2(fim2)
x = np.real(x)

##### Plotting the x slice
plt.figure()
plt.imshow(x, cmap=plt.cm.gray)
plt.axis('off')
```

```python
#plotting the y lines
# make a separate copy of image to work on
fim2 = fim.copy()
print(fim2.shape)
# set everything in all except the first row to zero
fim2[0,1:npy] = 0
# set everything in the first row to zero except the first 61 points and the last 60
fim2[61:npx-60,:] = 0

y = np.fft.ifft2(fim2)
y = np.real(y)

##### Plotting the x slice
plt.figure()
plt.imshow(y, cmap=plt.cm.gray)
plt.axis('off')
```
