```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import ensemble

np.random.seed(10)

def gen_sample_grid(npx=200, npy=200, limit = 4):
  x1line = np.linspace(-limit, limit, npx)
  x2line = np.linspace(-limit, limit, npy)
  x1grid, x2grid = np.meshgrid(x1line, x2line)
  Xgrid = np.array([x1grid, x2grid]).reshape([2,npx*npy]).T
  return Xgrid, x1line, x2line

Xgrid, x1line, x2line = gen_sample_grid()

from sklearn import datasets

X, y = datasets.make_classification(n_samples=10000,n_features=2,n_informative=2,n_redundant=0)

rf = ensemble.RandomForestClassifier(n_estimators=100,max_depth=20)
rf.fit(X,y)

prob = rf.predict_proba(Xgrid)
prob = np.reshape(prob[:,0],[200,200])

print(prob.shape)
plt.plot(x1line,prob[:,0])
```
